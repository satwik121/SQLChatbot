```python
import pandas as pd
import random

# Define the schema
data = {
    'user_id': [random.randint(1, 100) for _ in range(10)],
    'product_id': [random.randint(1000, 2000) for _ in range(10)],
    'product_name': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E',
                     'Product F', 'Product G', 'Product H', 'Product I', 'Product J'],
    'brand': ['Brand X', 'Brand Y', 'Brand Z', 'Brand X', 'Brand Y',
              'Brand Z', 'Brand X', 'Brand Y', 'Brand Z', 'Brand X'],
    'price': [random.uniform(10.0, 100.0) for _ in range(10)]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('my_database.db')

# Use the to_sql method to write the DataFrame to the database
df.to_sql('my_table', conn, if_exists='replace', index=False)

# Create a cursor object to interact with the database
cursor = conn.cursor()

# Execute an SQL query
query = "SELECT * FROM my_table"
cursor.execute(query)

# Fetch and print the results
results = cursor.fetchall()
for row in results:
    print(row)

# Close the cursor and the database connection
cursor.close()
conn.close()
```
